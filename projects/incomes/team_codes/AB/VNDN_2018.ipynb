```python
import pandas as pd
import numpy as np
from simpledbf import Dbf5
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
pd.options.display.float_format = "{:,.2f}".format
```

# Load data

```python
dbf = Dbf5('data/2018/HHOLD.dbf')
df = dbf.to_dataframe()
```

```python
df
```

# Рыночный доход (market income): все полученные доходы домохозяйства, без налогов и трансфертов

* (+) R_H_DOXOD1 (Т1_I) Доход от трудовой деятельности – всего

* (+) R_H_SOBST - К3_4 (Т1_7.3) Доходы от собственности (Включает в себя: - доход от полученных процентов со сбережений - доход от сдачи в аренду имущества - доход от сдачи в аренду земли - дивиденды по акциям и другим ценным бумагам)

* (+) R_H_TRANS_ (Т1_III) Трансферты – полученные

```python
df['market income'] = df['R_H_DOXOD1'] + df['R_H_SOBST'] + df['R_H_TRANS_']
```

# Располагаемый доход (disposable Income): рыночный доход - прямые налоги + прямые трансферты

* (+) R_H_DOX_RA - (Т1_VIII) Располагаемый совокупный доход (включая оценочный эквивалент чистой вмененной арендной платы)

```python
df['disposable income'] = df['R_H_DOX_RA']
```

# Нетрудовой доход (non-labor income)


* (+) R_H_SOBST - К3_4 (Т1_7.3) Доходы от собственности (Включает в себя: - доход от полученных процентов со сбережений - доход от сдачи в аренду имущества - доход от сдачи в аренду земли - дивиденды по акциям и другим ценным бумагам)

* (НЕ ВКЛЮЧАЕМ) R_H04_14_C - К3_5 (Т1_6.3.1) (Т1_7.2) Чистый доход от проживания в собственном жилье (чистая вмененная арендная плата), год 
* (+)  R_H_TRANS_ (Т1_III) Трансферты – полученные

```python
df['non-labor income'] = df['R_H_SOBST'] + df['R_H_TRANS_']
```
